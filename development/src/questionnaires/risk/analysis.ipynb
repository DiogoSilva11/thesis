# Analysis

```python
# import libraries

import json

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from scipy.stats import pearsonr, spearmanr
```

```python
def risk_distribution(data : dict, title="Risk Distribution") -> None:
    '''
    Displays a plot with risk distributions for the applicants
    '''
    plt.figure(figsize=(8, 4))
    sns.histplot(data, bins=10, color='green', edgecolor='black')
    plt.title(title, fontsize=16)
    plt.xlabel("Risk Score", fontsize=12)
    plt.ylabel("Frequency", fontsize=12)
    plt.tight_layout()
    plt.show()
```

```python
# load risk scores

data_folder = '../../../data/applicants'

with open(f"{data_folder}/scores.json", 'r') as file:
    scores = json.load(file)
```

```python
# score values

true_values = list(scores.values())
```

```python
# mortality risk distribution for all applicants

risk_distribution(true_values, "Risk Distribution (Ground Truth)")
```

```python
# basic statistics

scores_np = np.array(true_values)

mean = np.mean(scores_np)
median = np.median(scores_np)
mode = np.argmax(np.bincount(scores_np)) 
std_dev = np.std(scores_np)
min_score = np.min(scores_np)
max_score = np.max(scores_np)
percentiles = np.percentile(scores_np, [25, 50, 75]) 

print(f"Count: {len(scores)}")
print(f"Mean: {mean:.2f}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Standard Deviation: {std_dev:.2f}")
print(f"Min: {min_score}")
print(f"Max: {max_score}")
print(f"25th Percentile (Q1): {percentiles[0]}")
print(f"75th Percentile (Q3): {percentiles[2]}")
```

### Comparisons

```python
def superior_scores(bottom : float, top : float) -> int:
    '''
    Counts the number of occasions where dynamic performance was superior to traditional within a certain window
    '''
    count = 0
    for index in range(len(true_values)):
        truth = true_values[index]
        traditional = traditional_values[index]
        dynamic = dynamic_values[index]
        offset = (dynamic / truth) - (traditional / truth)
        if (truth - dynamic) <= (truth - traditional) and offset >= bottom and offset < top:
            count += 1
    return count
```

```python
def inferior_scores(bottom : float, top : float) -> int:
    '''
    Counts the number of occasions where dynamic performance was inferior to traditional within a certain window
    '''
    count = 0
    for index in range(len(true_values)):
        truth = true_values[index]
        traditional = traditional_values[index]
        dynamic = dynamic_values[index]
        offset = (traditional / truth) - (dynamic / truth)
        if (truth - dynamic) > (truth - traditional) and offset > bottom and offset <= top:
            count += 1
    return count
```

```python
# collect values from comparisons

superior_ranges = [(0.0, 0.1), (0.1, 0.2), (0.2, 0.3)]
superior_counts = [superior_scores(bottom, top) for bottom, top in superior_ranges]
superior_edges = [r[0] for r in superior_ranges] + [superior_ranges[-1][1]]
superior_x = np.arange(len(superior_counts))

inferior_ranges = [(0.0, 0.1), (0.1, 0.2), (0.2, 0.3), (0.3, 0.4), (0.4, 0.5), (0.5, 0.6), (0.6, 0.7), (0.7, 0.8)]
inferior_counts = [inferior_scores(bottom, top) for bottom, top in inferior_ranges]
inferior_edges = [r[0] for r in inferior_ranges] + [inferior_ranges[-1][1]]
inferior_x = np.arange(len(inferior_counts))

max_y = max(max(superior_counts), max(inferior_counts))
```

```python
# plot comparisons

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.bar(superior_x, superior_counts, color='green', edgecolor='grey', align='edge', width=1)
plt.title('Dynamic Equal or Superior to Traditional')
plt.xlabel('Improvement Range (%)')
plt.ylabel('Count')
plt.ylim(0, max_y + 1)
plt.xticks(np.arange(len(superior_edges)), [f"{int(e*100)}" for e in superior_edges])
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

plt.subplot(1, 2, 2)
plt.bar(inferior_x, inferior_counts, color='red', edgecolor='grey', align='edge', width=1)
plt.title('Dynamic Inferior to Traditional')
plt.xlabel('Decline Range (%)')
plt.ylabel('Count')
plt.ylim(0, max_y + 1)
plt.xticks(np.arange(len(inferior_edges)), [f"{int(e*100)}" for e in inferior_edges])
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
