# Analysis

```python
# import libraries

import json

import matplotlib.pyplot as plt
```

```python
# questionnaire sections

categories = [
    'personal',
    'lifestyle',
    'family',
    'health'
]
```

```python
# data folder

data_folder = '../../../data/questionnaires'
```

```python
# load questions

questions = {}
for category in categories:
    with open(f"{data_folder}/questions/{category}.json", 'r', encoding='utf-8') as f:
        questions[category] = json.load(f)
```

```python
# load stats

with open(f"{data_folder}/stats/traditional.json", 'r', encoding='utf-8') as file:
    stats = json.load(file)
```

```python
def factor_statistics(category: str, factor: str) -> None:
    '''
    Displays how much each option was selected for a specific factor
    '''
    if factor == 'personal_location' or factor == 'personal_occupation' or factor not in stats[category]:
        return
    question = questions[category][factor]['question']
    options = questions[category][factor]['options']
    answers = stats[category][factor]['answers']
    for index in range(len(options)):
        option = options[index]
        words = option.split(' ')
        if len(words) > 3:
            words = ' '.join(words[:3])
            options[index] = words + '...'
    plt.figure(figsize=(8, 4))
    plt.bar(options, answers, color='orange')
    plt.xlabel('Options')
    plt.ylabel('Count')
    plt.title(question)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

### Personal

```python
# display statistics for personal factors

category = 'personal'
for factor in questions[category].keys():
    factor_statistics(category, factor)
```

### Lifestyle

```python
# display statistics for lifestyle factors

category = 'lifestyle'
for factor in questions[category].keys():
    factor_statistics(category, factor)
```

### Family

```python
# display statistics for family factors

category = 'family'
for factor in questions[category].keys():
    factor_statistics(category, factor)
```

### Health

```python
# display statistics for health factors

category = 'health'
for factor in questions[category].keys():
    factor_statistics(category, factor)
```
