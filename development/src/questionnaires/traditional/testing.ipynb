```python
# import libraries

import json
```

```python
# data folder

data_folder = '../../../data/questionnaires'
```

```python
# questionnaire sections

categories = [
    'personal',
    'lifestyle',
    'family',
    'health'
]
```

```python
# load risks

risks = {}
for category in categories:
    with open(f"{data_folder}/questions/{category}.json", 'r', encoding='utf-8') as f:
        risks[category] = json.load(f)
```

```python
def questionnaire_section(category : str) -> list[dict]:
    '''
    Presents a set of questions for a section of the questionnaire
    '''
    section = []
    for index, content in enumerate(risks[category].values()):
        if index >= 10:
            break
        section.append({
            'questionId': index + 1,
            'question': content['question'],
            'options': content['options']
        })
    return section
```

```python
def display_section(section : list[dict]):
    '''
    Displays the contents of a questionnaire section
    '''
    for item in section:
        print('-' * 120)
        print(item['question'])
        for index, option in enumerate(item['options']):
            print(f"{index}. {option}")
```

```python
# personal section

section = questionnaire_section('personal')
display_section(section)
```

```python
# lifestyle section

section = questionnaire_section('lifestyle')
display_section(section)
```

```python
# family section

section = questionnaire_section('family')
display_section(section)
```

```python
# health section

section = questionnaire_section('health')
display_section(section)
```
