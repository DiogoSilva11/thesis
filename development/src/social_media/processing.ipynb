# Processing

```python
# import libraries

import json
```

```python
# data folder

data_folder = '../../data/social_media'
```

```python
# load all captions

with open(f"{data_folder}/captions.json", 'r', encoding='utf-8') as file:
    captions = json.load(file)
```

```python
# indian term conversions

terms = {
    'in india': 'in portugal',
    'of indian': 'of',
    'an indian': 'a',
    'the indian': 'the',
    'traditional indian': 'traditional',
    'cricket': 'football',
    'taking the wicket': 'scoring a goal',
    ' sari ': ' dress ',
    'red and gold sari': 'red and gold dress',
    'saris': 'dresses' 
}
```

```python
# formatting issues conversion

issues = {
    "s '": "s'",
    " ' s": "'s",
    " ' t": "'t",
    "' ralph": "ralph",
    "kingdom '": "kingdom",
    " %": "%",
    " ' re": "'re"
}
```

```python
# process all captions - remove first part and replace indian terms or formatting issues 

processed_captions = []

for caption in captions:
    processed_caption = caption
    processed = False
    for key, value in terms.items():
        if key in caption:
            processed = True
            processed_caption = processed_caption.replace(key, value)
    for key, value in issues.items():
        if key in caption:
            processed = True
            processed_caption = processed_caption.replace(key, value)
    processed_caption = processed_caption[len('a photograph of '):]
    processed_captions.append(processed_caption)
    if processed:
        print('-' * 120)
        print(caption)
        print(processed_caption)
```

```python
# save processed captions

with open(f"{data_folder}/processed_captions.json", 'w', encoding='utf-8') as file:
    json.dump(processed_captions, file, indent=4)
```
